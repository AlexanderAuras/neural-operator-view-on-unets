```python
import numpy as np
from numpy import fft
import matplotlib.pyplot as plt
from covariance import generate_covmatrix
```

```python
N = 512
white_noise = np.random.randn(N,N)
decay_filter = generate_covmatrix(N)
decay_noise = fft.ifft2(fft.fft2(white_noise)*decay_filter).real

fig, axs = plt.subplots(1, 3, figsize = (10,15))
axs[0].imshow(white_noise)
axs[0].set_title("White noise")
axs[1].imshow(np.log(decay_filter))
axs[1].set_title("log(Filter)")
axs[2].imshow(decay_noise)
axs[2].set_title("Trace-class noise")
plt.show()
```
