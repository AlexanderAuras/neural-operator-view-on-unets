```python
import matplotlib.pyplot as plt
import torch

from fno_unet.noise_filter import generate_noise_filter
```

```python
N = 512
white_noise = torch.randn(N,N)
decay_filter = generate_noise_filter(N)
decay_noise = torch.fft.ifft2(torch.fft.fft2(white_noise, norm="ortho")*decay_filter, norm="ortho").real

fig, axs = plt.subplots(1, 3, figsize = (10,15))
axs[0].imshow(white_noise)
axs[0].set_title("White noise")
axs[1].imshow(torch.log(decay_filter))
axs[1].set_title("log(Filter)")
axs[2].imshow(decay_noise)
axs[2].set_title("Trace-class noise")
plt.show()
```
